```python
filePath = "/content/100-0.txt"
ShakespeareFile = open(filePath, "r")
ShakespeareContent = ShakespeareFile.read()
```

```python
ShakespeareContent
```

**TASK 2.1**

```python
ShakespeareContent_List = ShakespeareContent.splitlines()  # The splitlines() method splits a string into a list. The splitting is done at line breaks.
print(ShakespeareContent_List[0:50])
```

**TASK 2.2**

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(ShakespeareContent_List)
word_counts = tokenizer.word_counts
word_index = tokenizer.word_index

num_tokens = sum(word_counts.values())
num_types = len(word_index)
max_value = max(word_counts, key=word_counts.get)

print(f"Number of Tokens: {num_tokens}")
print(f"Number of Types: {num_types}")
print(f"Most frequent Type is: '{max_value}' with {word_counts[max_value]} occurances")
```

**TASK 2.3**

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

tokens = list(word_index.keys())
stemmed_types = set([stemmer.stem(token) for token in tokens])
lemmatized_types = set([lemmatizer.lemmatize(token) for token in tokens])

number_of_stemmed_types = len(stemmed_types)
number_of_lemmatized_types = len(lemmatized_types)

# Asserting the validity of the expression
try:
    assert num_types > number_of_lemmatized_types > number_of_stemmed_types, 'Expression validation failed'
except Exception as e:
    print(f"Assertion failed: {e}")

print(f"Number of Types: {num_types}")
print(f"Number of Types after Stemming: {number_of_stemmed_types}")
print(f"Number of Types after Lemmatization: {number_of_lemmatized_types}")
```

**TASK 2.4**

```python
import spacy

spacy_nlp = spacy.load("en_core_web_sm")

last_100_lines = ShakespeareContent.splitlines()[-100:]
last_100_lines = " ".join(last_100_lines)

doc = spacy_nlp(last_100_lines)
segmented_sentences = []

for sentence in doc.sents:
    segmented_sentences.append(sentence.text)

print("Segmented sentences:")
for i, sentence in enumerate(segmented_sentences, start=1):
    print(f"{i}. {sentence}")

print("--" * 100)
print("Total number of segmented sentences:", len(segmented_sentences))
```
