```python
import nltk
from nltk import bigrams, FreqDist
from nltk.tokenize import word_tokenize

def probability(sentence, unigram_counts, bigram_counts):
    words = sentence.split()
    sentence_probability = 1.0

    for i in range(len(words) - 1):
        word = words[i]
        next_word = words[i + 1]

        # Calculate unigram count for the current word
        unigram_count = unigram_counts.get(word, 0)

        # Calculate bigram count for the current word pair
        bigram_count = bigram_counts.get((word, next_word), 0)

        # Calculate the bigram probability for the current word pair
        if unigram_count > 0:
            bigram_probability = bigram_count / unigram_count
        else:
            # Handle the case where the unigram count is 0
            bigram_probability = 0

        sentence_probability *= bigram_probability

        print(f"unigram count for '{word}' = {unigram_count}")
        print(f"bigram count for '{word} {next_word}' = {bigram_count}")
        print(f"bigram probability for '{word} {next_word}' = {bigram_probability}\n")

    return sentence_probability
```

```python
sentence1 = "i love you"
sentence2 = "i hate you"

probability1 = probability(sentence1, unigrams_df_dict, bigrams_df_dict)
print()
probability2 = probability(sentence2, unigrams_df_dict, bigrams_df_dict)

print(f"\nSentence probability for '{sentence1}' = {probability1}")
print(f"Sentence probability for '{sentence2}' = {probability2}\n")

print(probability1 > probability2)
```

```python
import random

def ShannonVisualization(bigram_df_dict, start_word="<S>", max_iter=float('inf')):
    output = []  # Initialize the output as a list of words
    current_word = start_word
    iter_count = 0

    while current_word != "</S>":
        # Retrieve bigram probabilities for the current word
        bigram_probabilities = {bigram: count for bigram, count in bigram_df_dict.items() if bigram[0] == current_word}
        # print('bigram_probabilities: ', bigram_probabilities)

        # If no available bigrams, break the loop
        if not bigram_probabilities:
            break

        # Calculate the total count of bigrams starting with the current word
        total_count = sum(bigram_probabilities.values())
        # print('total_count: ', total_count)

        # Calculate the normalized probabilities for the next bigrams
        next_bigrams = list(bigram_probabilities.keys())
        # print('next_bigrams: ', next_bigrams)

        probabilities = [count / total_count for count in bigram_probabilities.values()]
        # print('probabilities: ', probabilities)

        # Choose the next bigram based on probabilities
        next_bigram = random.choices(next_bigrams, weights=probabilities)[0]

        # Uncomment the below line if we want to pick the next word based on maximum probability
        # next_bigram = max(next_bigrams, key=lambda bigram: bigram_probabilities[bigram])

        # print('next_bigram: ', next_bigram)

        # Append the second word of the bigram to the output
        if next_bigram[1] != '</S>': output.append(next_bigram[1])

        current_word = next_bigram[1]  # Set the current word to the second word of the chosen bigram
        # print('\n')

        if iter_count == max_iter:
          break
        iter_count += 1

    return " ".join(output)
```
