```python
import pandas as pd
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier

import nltk
nltk.download('punkt')
```

```python
def trainAndTestClassifier(trainingSet, testSet, flag=0):
    classifier = NaiveBayesClassifier(trainingSet)
    if flag == 1:
      classification_list = [(text, classifier.classify(text[0])) for text in testSet]
      for text, label in classification_list:
          print(f'Document: {text}, Predicted Class: {label}')
      print()
    accuracy = classifier.accuracy(testSet)
    print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
trainingSet = [
    ('London is the Capital of GB', 'GB'),
    ('Oxford is a city in GB', 'GB'),
    ('Dublin is the capital of Ireland', 'IE'),
    ('Limerick is a city in Ireland', 'IE')
]

testSet = [
    ('University of Limerick', 'IE'),
    ('University College Dublin', 'IE'),
    ('Imperial College London', 'GB'),
    ('University of Oxford', 'GB'),
    ('Ireland & GB', 'GB')
]

trainAndTestClassifier(trainingSet, testSet, 1)
```

**Trial with K Fold approach**

```python
from sklearn.model_selection import KFold

df = pd.read_csv('/content/bbc-text.csv')

kf = KFold(n_splits=10, shuffle=True, random_state=42)

accuracy_scores = []
iter = 1

for train_index, test_index in kf.split(df):
    train_set = df.loc[train_index]
    test_set = df.loc[test_index]

    training_set_tuples = list(zip(train_set['text'], train_set['category']))
    test_set_tuples = list(zip(test_set['text'], test_set['category']))

    classifier = NaiveBayesClassifier(training_set_tuples)
    accuracy = classifier.accuracy(test_set_tuples)
    accuracy_scores.append(accuracy)
    print(f'Iter: {iter} :: Accuracy: {accuracy}')

    del classifier
    del train_set
    del test_set
    iter += 1

print()
average_accuracy = sum(accuracy_scores) / len(accuracy_scores)
print(f'Average accuracy over 10 Folds: {average_accuracy * 100:.2f}%')
```
