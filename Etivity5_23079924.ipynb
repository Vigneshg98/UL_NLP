```python
import pandas as pd
```

```python
column_names = ['word']
positive_df = pd.read_csv("/content/positive-words.txt", sep='\t', header=None, names=column_names, skiprows=34)
negative_df = pd.read_csv("/content/negative-words.txt", sep='\t', header=None, names=column_names, encoding='latin-1', skiprows=34)
```

```python
display(positive_df.tail(5))

display(negative_df.tail(5))
```

```python
positive_list = positive_df['word'].values.tolist()
negative_list = negative_df['word'].values.tolist()
```

```python
def sentimentAnalyzer(inp_string):
    words = inp_string.lower().split()

    positive_count = sum(word in positive_list for word in words)
    negative_count = sum(word in negative_list for word in words)
    total_words = len(words)

    sentiment = "Positive Sentiment" if positive_count > negative_count else "Negative Sentiment" if negative_count > positive_count else "Neutral Sentiment"
    confidence = max(positive_count, negative_count) / total_words if total_words > 0 else 0.0

    return sentiment, confidence
```

```python
inp_string_list = [
    "NLP is cool",
    "NLP is cool and useful",
    "NLP is hard",
    "NLP is hard and useless",
    "NLP stands for Natural Language Processing"
]
for inp_string in inp_string_list:
  sentiment, confidence = sentimentAnalyzer(inp_string)
  print("String: ", inp_string)
  print("Sentiment:", sentiment)
  print("Confidence:", confidence)
  print("")
```
