```python
file_path = "/content/Eircode Routing Key Boundaries.csv"
```

```python
# Reading the csv file and converting it as a dictionary for easy access
with open(file_path, 'r') as f:
  reader = csv.reader(f)
  rounting_key_dict = dict(list(map(tuple, reader)))
```

```python
list(rounting_key_dict.items())[:5]
```

```python
def eirCodeValidator(eircode):
  print('Eircode: ', eircode)
  eircode = eircode.strip()
  flag = True
  routing_key = ''
  unique_identifier = ''
  descriptor = ''

  # Check if the eircode matches the global pattern of eircode
  flag = bool(re.match('^[A-Z0-9]{3}\s?[-]?[A-Z0-9]{4}$', eircode))

  if flag:
    eircode = eircode.replace('-', ' ')
    # Tokenize the eircode using nltk library
    tokens = nltk.word_tokenize(eircode)
    print('Tokens: ', tokens)

    # Get the first part of eircode
    routing_key = tokens[0]

    # Get the second part of eircode
    unique_identifier = tokens[1]

    flag = False
    if routing_key in rounting_key_dict:
      flag = True
      descriptor = rounting_key_dict[routing_key]

  return flag, routing_key, unique_identifier, descriptor
```

```python
sample_eircodes = ['V94-TP9X', 'V94 TP9X', 'v94 TP9X', ' V94 TP9X', 'V94 TP9X ', 'A65 F4E2', 'D01 ABCD']

for eircode in sample_eircodes:
  flag, routing_key, unique_identifier, descriptor = eirCodeValidator(eircode)
  print('Valid Eircode' if flag else 'Invalid Eircode')
  if flag:
    print(f'Routing Key: {routing_key} :: Unique Identifier: {unique_identifier} :: Destination: {descriptor}')
  print('-'*10)
```
