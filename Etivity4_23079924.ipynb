```python
import pandas as pd
import jellyfish
from textblob import TextBlob, Word
```

```python
column_names = ['word', 'freq']
unigrams_df = pd.read_csv("/content/count_1w.txt", sep='\t', header=None, names=column_names)
```

```python
display(unigrams_df.head(5))
```

```python
unigrams_df_dict = unigrams_df.set_index('word')['freq'].to_dict()
```

```python
def nonWordSpellingCorrection(input_word, unigrams_df, max_distance=1):
    suggestions = []
    total = sum(unigrams_df['freq'])
    for _, row in unigrams_df.iterrows():
        try:
          distance = jellyfish.damerau_levenshtein_distance(input_word, row['word'])
          if distance <= max_distance:
              suggestions.append({
                  "word": row['word'],
                  "frequency": row['freq'],
                  "probability": row['freq'] / total
              })
        except:
          pass
    candidates_df = pd.DataFrame(suggestions)
    candidates_df = candidates_df.sort_values(by=["probability"], ascending=False)
    return candidates_df
```

```python
inp_word = 'acress'
# max_distance = 2
# pass max_distance to desired value as the third parameter to the nonWordSpellingCorrection(), else default max_distance will be 1
candidates_df = nonWordSpellingCorrection(inp_word, unigrams_df)
print('nonWord: ', inp_word)
display(candidates_df)
```

```python
def getSpellingSuggestions(inp_word):
  result = []
  word = Word(inp_word)
  suggestions = word.spellcheck()
  for suggestion, probability in suggestions:
    result.append((suggestion, probability))
  return result

inp_word = 'acress'
print('Word: ', inp_word)
print('Candidates: ', getSpellingSuggestions(inp_word))
```
