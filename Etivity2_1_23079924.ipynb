```python
import time

def edDistRecursive(x, y):

    def customSubstitionCost(str1, str2):
        return 2 if (str1 != str2) and ((str1.isalpha() and str2.isnumeric()) or (str2.isalpha() and str1.isnumeric())) else 1 if str1 != str2 else 0

    if len(x) == 0:
        #print(f'\t Converting "{x}" TO "{y}" requires {len(y)} INSERTIONS, therefor it costs {len(y)}')
        return len(y)
    if len(y) == 0:
        #print(f'\t Converting "{x}" TO "{y}" requires {len(x)} DELETIONS, therefor it costs {len(x)}')
        return len(x)

    delta = customSubstitionCost(x[-1], y[-1])
    diagonal_or_substitution_cost = edDistRecursive(x[:-1], y[:-1]) + delta
    vertical_or_deletion_cost = edDistRecursive(x[:-1], y) + 1
    horizontal_or_insertion_cost = edDistRecursive(x, y[:-1]) + 1

    minValue = min(diagonal_or_substitution_cost, vertical_or_deletion_cost, horizontal_or_insertion_cost)
    return minValue
```

```python
start_time = time.time()
x = "intention"
y = "execution"
distance = edDistRecursive(x, y)
end_time = time.time()

print(f"Edit Distance between '{x}' and '{y}': {distance}")
print(f"Executed in: {end_time - start_time} seconds")
```

```python
from names_dataset import NameDataset, NameWrapper
import jellyfish
```

```python
class PlayWithJellyFish(object):

  def __init__(self):
    self.nd = NameDataset()

  def checkIfNamesExist(self, first_name, last_name):
      fn = self.nd.search(first_name)
      ln = self.nd.search(last_name)
      first_name_exists = True if fn['first_name'] else False
      last_name_exists = True if ln['last_name'] else False
      return first_name_exists, last_name_exists

  def getSoundexCode(self, first_name, last_name):
      first_name_soundex = jellyfish.soundex(first_name)
      last_name_soundex = jellyfish.soundex(last_name)
      return first_name_soundex, last_name_soundex

  def getMatchingSoundexFirstNames(self, first_name):
      input_soundex = jellyfish.soundex(first_name)
      matching_names = [name for name in self.nd.first_names if jellyfish.soundex(name) == input_soundex]
      return matching_names

  def getLevenshteinDistance(self, first_name, last_name):
      distance = jellyfish.levenshtein_distance(first_name, last_name)
      return distance
```

```python
obj = PlayWithJellyFish()
```

```python
input_first_name = "Albert"
input_last_name = "Einstein"

# print(NameWrapper(obj.nd.search(input_first_name)).describe)
# print(NameWrapper(obj.nd.search(input_last_name)).describe)

# Task A
first_name_exists, last_name_exists = obj.checkIfNamesExist(input_first_name, input_last_name)
print(f"First Name exists: {first_name_exists}")
print(f"Last Name exists: {last_name_exists}")
print("")

# Task B
first_name_soundex, last_name_soundex = obj.getSoundexCode(input_first_name, input_last_name)
print(f"Soundex Code for First Name: {first_name_soundex}")
print(f"Soundex Code for Last Name: {last_name_soundex}")
print("")

# Task C
matching_names = obj.getMatchingSoundexFirstNames(input_first_name)
print(f"First Names with the Same Soundex Code as '{input_first_name}': {matching_names}")
print("")

# Task D
levenshtein_distance = obj.getLevenshteinDistance(input_first_name, input_last_name)
print(f"Levenshtein Distance between '{input_first_name}' and '{input_last_name}': {levenshtein_distance}")
```
